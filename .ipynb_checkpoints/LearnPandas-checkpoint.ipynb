```python
#Verify if pandas is installed

import pandas as pd
# if there are no errors then pandas is installed successfully.
```

```python
# Components - Series and DataFrame

#Series - is a column in pandas.

#DataFrame - is a table which can have multiple columns (or you can say a collection of Series).
#DataFrame can be created as a dictionary. Keys in dictionary turn into columns and values turn to column values.

players_score = {'Sachin' : [99, 81, 56, 122], 'Rahul' : [101, 88, 92, 126]}
players_df = pd.DataFrame(players_score)

players_df
```

```python
#Adding indexes
players_df = pd.DataFrame(players_score, index = [2009, 2010, 2011, 2012])
players_df
```

```python
#Location data in DataFrame
players_df.loc[2011]
```

```python
#reading data from external source to pandas
#use read_json for json formatted data, 
#use pysqlite3 to read from a database via sql select query using read_sql_query

#Reading a csv file in pandas 
df_file = pd.read_csv('data/SalesRecords.csv', index_col=0)
df_file
```

```python
#Convert from DataFrame to csv, json or Sql
#to_csv, to_json, to_sql
```

```python
#Viewing data
df_file.head() #returns the first 5 rows
df_file.tail() #returns last 5 rows
```

```python
#Details about data
df_file.info() #no of rows, non null values and data type etc
```

```python
#Get no of rows and columns
df_file.shape
```

```python
df_temp = df_file.append(df_file) #append returns a copy without modifying the actual df_file
df_temp.shape

df_temp = df_temp.drop_duplicates() #returns a copy of DataFrame

df_temp.shape

```

```python
#inplace=True modifies the DataFrame object in place

#Keep has three options first, last and False
# first - default, drop duplicates except for the first occurence
# last - drop duplicates except fot the last occurence
# False - Drop all duplicates

df_temp.drop_duplicates(inplace=True, keep=False)
df_temp.shape
```

```python
#Read columns

df_file.columns
```

```python
#renaming columns

df_file.rename(columns={'Country' : 'CountryName'}, inplace=True)
df_file

#lowercase columns
df_file.columns = [col.lower() for col in df_file]
df_file.columns
```

```python
#Handling missing values
df_file.isnull()

df_file.isnull().sum() #returns the number of nulls in each column
```

```python
#Null values can either be deleted or replace with some other value mean or median
df_file.dropna() #removes null, deletes any row with atleast one null value
```

```python
#df_file.dropna(axis=1) #removes columns containing null values
```

```python
total_cost = df_file['total cost']
total_cost
```

```python
#Calculate mean of column total cost
total_cost_mean = total_cost.mean()
total_cost_mean
```

```python
#Fill the nulls using mean
total_cost.fillna(total_cost_mean, inplace=True)
```

```python
#verify that total_cost doesn't have null value
df_file.isnull().sum()
```

```python
df_file.describe()
df_file['countryname'].describe() #top value of a given column
```

```python
#frequency of values in a column
df_file['countryname'].value_counts().head()
```

```python
#Relationship between continous variables using correlation method
#Positive and negavtive correlation.
#Posiitve correlation between two variables indicates that if one variable goes up, the other goes up as well.
#Negative correlation between two variables indicates that if one variables goes up, the other goes down.
df_file.corr()
```

```python
#Conditional selection of data

data_conditional = df_file[df_file['countryname'] == "Russia"] #Select dataframe where countryname is Russian
data_conditional.head()
```

```python
data_conditional2 = df_file[df_file['total revenue'] >= 5000000 ].head()
data_conditional2
```

```python
#checking & or | condition
df_file[(df_file['countryname'] == 'Lithuania') & (df_file['units sold'] >= 500)]
```

```python
# checking in condition
df_file[df_file['countryname'].isin(['Russia', 'Myanmar', 'Belgium'])].head()
```

```python
#Applying functions to dataset


def rating_sales(x):
    if x >= 5000:
        return 'Good'
    else:
        return 'Poor'
    

df_file['sales performance'] = df_file['units sold'].apply(rating_sales)
#Apply function passes every value from 'units sold' column to rating_sales function and the result is then returned as a series.
# This series is then assigned to a new column called 'sales performance'

#Apply function will be much faster than iterating manually over rows because pandas does it via vectorization.

df_file.head()
    
```
