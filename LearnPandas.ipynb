```python
#Verify if pandas is installed

import pandas as pd
# if there are no errors then pandas is installed successfully.
```

```python
# Components - Series and DataFrame

#Series - is a column in pandas.

#DataFrame - is a table which can have multiple columns (or you can say a collection of Series).
#DataFrame can be created as a dictionary. Keys in dictionary turn into columns and values turn to column values.

players_score = {'Sachin' : [99, 81, 56, 122], 'Rahul' : [101, 88, 92, 126]}
players_df = pd.DataFrame(players_score)

players_df
```

```python
#Adding indexes
players_df = pd.DataFrame(players_score, index = [2009, 2010, 2011, 2012])
players_df
```

```python
#Location data in DataFrame
players_df.loc[2011]
```

```python
#reading data from external source to pandas
#use read_json for json formatted data, 
#use pysqlite3 to read from a database via sql select query using read_sql_query

#Reading a csv file in pandas 
df_file = pd.read_csv('data/SalesRecords.csv', index_col=0)
df_file
```

```python
#Convert from DataFrame to csv, json or Sql
#to_csv, to_json, to_sql
```

```python
#Viewing data
df_file.head() #returns the first 5 rows
df_file.tail() #returns last 5 rows
```

```python
#Details about data
df_file.info() #no of rows, non null values and data type etc
```

```python
#Get no of rows and columns
df_file.shape
```

```python
df_temp = df_file.append(df_file) #append returns a copy without modifying the actual df_file
df_temp.shape

df_temp = df_temp.drop_duplicates() #returns a copy of DataFrame

df_temp.shape

```

```python
#inplace=True modifies the DataFrame object in place

#Keep has three options first, last and False
# first - default, drop duplicates except for the first occurence
# last - drop duplicates except fot the last occurence
# False - Drop all duplicates

df_temp.drop_duplicates(inplace=True, keep=False)
df_temp.shape
```

```python
#Read columns

df_file.columns
```

```python
#renaming columns

df_file.rename(columns={'Country' : 'CountryName'}, inplace=True)
df_file
```
